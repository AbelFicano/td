```python
import tdwf

ad2 = tdwf.AD2()
scope = tdwf.Scope(ad2.hdwf)
scope.fs = 1e6
scope.npt = 1000

scope.ch1.rng = 5
scope.ch2.rng = 50
scope.ch2.avg = True

scope.sample()

import matplotlib.pyplot as plt

plt.plot(scope.time.vals, scope.ch1.vals)
plt.xlabel("Tempo [s]")
plt.ylabel("Ch1 [V]")

plt.savefig('Figura 3')
```
