```python
import tdwf
import numpy as np
import matplotlib.pyplot as plt

ad2 = tdwf.AD2()
```

```python
wgen = tdwf.WaveGen(ad2.hdwf)
wgen.w1.offs = 2
wgen.w1.func = tdwf.funcCustom
wgen.w1.data = np.linspace(-1,1,10000)
wgen.w1.ampl = 2
wgen.w1.freq = 1e6/8192*2
wgen.w1.start()
```

```python
scope = tdwf.Scope(ad2.hdwf)
scope.fs = 1e6
scope.npt = 8192
```

```python
scope.ch1.offs = 3
scope.ch1.rng = 5
scope.ch1.avg = True
scope.ch2.offs = 3
scope.ch2.rng = 5
scope.ch2.avg = True
```

```python
scope.sample()
```

```python
scope.ch1.vals
```

```python
scope.ch2.vals
```

```python
data = np.column_stack((scope.time.vals,scope.ch1.vals,scope.ch2.vals))
np.savetxt("dati.txt", data, delimiter="\t")
```

```python
plt.plot(scope.time.vals, scope.ch1.vals)
plt.xlabel("Tempo [s]")
plt.ylabel("Ch1 [V]")
```

```python
plt.plot(scope.time.vals, scope.ch2.vals)
plt.xlabel("Tempo [s]")
plt.ylabel("Ch2 [V]")
```
