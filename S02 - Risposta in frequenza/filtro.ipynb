```python
import tdwf
import numpy as np
import matplotlib.pyplot as plt
```

```python
ad2 = tdwf.AD2()
```

```python
wgen = tdwf.WaveGen(ad2.hdwf)
wgen.w1.start()
```

```python
scope = tdwf.Scope(ad2.hdwf)
scope.fs = 1e6
scope.npt = 1000

scope.ch1.rng = 5
scope.ch1.avg = True
scope.ch2.rng = 5
scope.ch2.avg = True
```

```python
scope.sample()
```

```python
data = np.column_stack((scope.time.vals,scope.ch1.vals,scope.ch2.vals))
```

```python
plt.plot(data[:,0], data[:,1])
plt.xlabel("Tempo [s]")
plt.ylabel("Ch [V]")
```

```python
plt.plot(data[:,0], data[:,2])
plt.xlabel("Tempo [s]")
plt.ylabel("Ch [V]")
```
