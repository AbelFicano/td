```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
banda = np.loadtxt('passabanda.txt')
basso = np.loadtxt('passabasso.txt')
alto = np.loadtxt('passaalto.txt')
```

```python
banda
```

```python
plt.loglog(banda[:,0],banda[:,1])
plt.loglog(banda[:,0],basso[:,1]*alto[:,1])
plt.legend(['Passa banda','Prodotto'])
plt.show()
```

```python
plt.semilogx(banda[:,0],banda[:,3])
plt.semilogx(banda[:,0],basso[:,3]+alto[:,3])
plt.legend(['Passa banda','Prodotto'])
plt.show()
```
