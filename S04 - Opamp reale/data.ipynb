```python
import numpy as np
import tdwf
import sweepfreq
```

```python
f0=1e2
f1=1e6
nf=40
npt = 8192

# -[Configurazione AD2]--------------------------------------------------------
#   1. Connessiene con AD2 e selezione configurazione 
ad2 = tdwf.AD2()
#   2. Configurazione generatore di funzioni
wavegen = tdwf.WaveGen(ad2.hdwf)
wavegen.w1.ampl = 2
wavegen.w1.start()

# Configurazione alimentazione
ad2.vdd = +5
ad2.vss = -5
ad2.power(True)

#   3. Configurazione oscilloscopio
scope = tdwf.Scope(ad2.hdwf)
scope.fs = 1e6
scope.npt = npt
scope.ch1.rng = 5
scope.ch2.rng = 5

sweepfreq.sweep(wavegen, scope, f0, f1, nf, npt, n=4, filename='dataAD8031_330.txt')

ad2.close()
```
